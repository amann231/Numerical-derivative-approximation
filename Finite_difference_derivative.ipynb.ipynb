## Theory: Finite Differences

### What is a Gradient/Slope?

In simple terms, the **gradient** or **slope** of a function at a particular point tells us how steeply the function's value is changing at that point. Imagine walking on a landscape; the slope at your current position indicates how steep the ground is right where you are standing. A positive slope means the function is increasing, a negative slope means it's decreasing, and a zero slope means it's momentarily flat.

Mathematically, the slope of a function `f(x)` at a point `x` is called its **derivative**, denoted as `f'(x)` or `dy/dx`.

### Numerical Differentiation

Sometimes, finding the exact analytical derivative of a function can be complex or impossible, especially when we only have a set of discrete data points rather than a continuous function. **Numerical differentiation** provides a way to approximate this derivative using known function values.

We will explore two common methods for numerical differentiation: Forward Finite Difference and Backward Finite Difference.

---

### Forward Finite Difference

The **Forward Finite Difference** method approximates the derivative at a point `x` by looking at the function's value at `x` and a small step `h` *forward* from `x` (i.e., `x + h`).

**Conceptual Explanation:**
Imagine drawing a straight line (a secant line) between the point `(x, f(x))` and the point `(x + h, f(x + h))` on the function's curve. The slope of this secant line is used as an approximation for the derivative at `x`.

**Mathematical Formula:**

`f'(x) \approx \frac{f(x + h) - f(x)}{h}`

---

### Backward Finite Difference

The **Backward Finite Difference** method approximates the derivative at a point `x` by looking at the function's value at `x` and a small step `h` *backward* from `x` (i.e., `x - h`).

**Conceptual Explanation:**
Similar to the forward difference, we draw a secant line, but this time between the point `(x - h, f(x - h))` and the point `(x, f(x))` on the function's curve. The slope of this secant line approximates the derivative at `x`.

**Mathematical Formula:**

`f'(x) \approx \frac{f(x) - f(x - h)}{h}`

---

### The Role of `h` (Step Size)

The parameter `h` (the step size) is crucial in both forward and backward finite difference methods. It represents the small increment or decrement from `x` used to calculate the approximation.

*   **Impact on Accuracy:** Generally, a smaller `h` leads to a more accurate approximation of the true derivative because the secant line gets closer to the tangent line at `x`. However, `h` cannot be infinitely small. If `h` is too small, numerical precision issues (e.g., floating-point errors) can arise, leading to increased error.
*   **Impact on Computational Cost:** A smaller `h` usually doesn't significantly impact computational cost for these simple methods, but it's a key factor in more complex numerical methods.

## Implementing Forward Finite Difference



**Reasoning**:
I will define the `forward_difference` function, calculate the step size `h` from the generated `x` values, apply the forward difference method to the `x` values (excluding the last one), store the results in `f_prime_forward`, and then print the first few values to verify the implementation.



```python
import numpy as np

# 1. Define the forward_difference function
def forward_difference(f, x_val, h):
    return (f(x_val + h) - f(x_val)) / h

# 2. Choose an appropriate step size h
# Since x was generated using np.linspace, the step size is constant
h = x[1] - x[0]

# 3. Apply the forward_difference function to the x values (excluding the last one)
f_prime_forward = np.array([forward_difference(f, x_val, h) for x_val in x[:-1]])

# 4. Store the calculated forward differences in f_prime_forward (already done above)

# 5. Print the first few calculated forward differences
print("First 5 calculated forward differences:")
print(f_prime_forward[:5])
print("Last 5 calculated forward differences:")
print(f_prime_forward[-5:])
```

## Implementing Backward Finite Difference



**Reasoning**:
I will define the `backward_difference` function, apply it to the appropriate `x` values (excluding the first one) using the previously calculated step size `h`, store the results in `f_prime_backward`, and then print the first and last few values to verify the implementation.



```python
import numpy as np

# Assuming f(x), x, and h are already defined from previous steps

# 1. Define the backward_difference function
def backward_difference(f, x_val, h):
    return (f(x_val) - f(x_val - h)) / h

# 2. Use the previously calculated step size h (already available)
# h = x[1] - x[0]

# 3. Apply the backward_difference function to the x values (excluding the first one)
f_prime_backward = np.array([backward_difference(f, x_val, h) for x_val in x[1:]])

# 4. Store the calculated backward differences in f_prime_backward (already done above)

# 5. Print the first few and last few calculated backward differences
print("First 5 calculated backward differences:")
print(f_prime_backward[:5])
print("Last 5 calculated backward differences:")
print(f_prime_backward[-5:])
```

## Calculate Analytical Derivative

### Explanation of Analytical Derivative for `f(x) = x^2`

The analytical derivative of a function gives us its exact instantaneous rate of change at any given point. For the chosen function `f(x) = x^2`, we can derive its analytical derivative using the basic rules of differentiation.

**Power Rule of Differentiation**

The most fundamental rule applicable here is the **power rule**, which states that if `f(x) = x^n`, then its derivative `f'(x)` is `n * x^(n-1)`.

Let's apply this to `f(x) = x^2`:

1.  **Identify `n`**: In our function `f(x) = x^2`, the exponent `n` is `2`.
2.  **Apply the Power Rule**: According to the power rule, `f'(x) = n * x^(n-1)`.
    *   Substitute `n = 2` into the formula:
        `f'(x) = 2 * x^(2-1)`
    *   Simplify the exponent:
        `f'(x) = 2 * x^1`
    *   Which simplifies to:
        `f'(x) = 2x`

Therefore, the analytical (or exact) derivative of `f(x) = x^2` is `f'(x) = 2x`. This exact solution will serve as the benchmark against which we will compare the approximations from the finite difference methods.

## Perform Comparison and Error Analysis


```python
import numpy as np

# 1. Calculate the absolute error for forward finite difference
# f_prime_forward corresponds to x[:-1], so compare with analytical_derivative(x[:-1])
error_forward = np.abs(f_prime_forward - f_prime_analytical[:-1])

# 2. Calculate the absolute error for backward finite difference
# f_prime_backward corresponds to x[1:], so compare with analytical_derivative(x[1:])
error_backward = np.abs(f_prime_backward - f_prime_analytical[1:])

# 3. Print the first few and last few values of error_forward and error_backward
print("First 5 calculated forward error values:")
print(error_forward[:5])
print("Last 5 calculated forward error values:")
print(error_forward[-5:])
print("\nFirst 5 calculated backward error values:")
print(error_backward[:5])
print("Last 5 calculated backward error values:")
print(error_backward[-5:])
```

### Explanation of Absolute Error

In this context, **absolute error** quantifies the magnitude of the difference between our numerical approximations (forward and backward finite differences) and the exact analytical derivative. It is calculated as:

`Absolute Error = |Numerical Derivative - Analytical Derivative|`

*   A **smaller absolute error** indicates that the numerical method provides a more accurate approximation of the true derivative.
*   A **larger absolute error** suggests a less accurate approximation.

By calculating the absolute error for both forward and backward finite difference methods, we can directly compare their performance and understand how well each method approximates the true derivative for our chosen function `f(x) = x^2` with the given step size `h`. The uniformity of the absolute errors across the `x` range for both methods suggests that for `f(x)=x^2`, which has a constant second derivative, the error of both first-order methods (forward and backward) is constant and directly proportional to the step size `h`.
